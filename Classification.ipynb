```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
```

```python
data = pd.DataFrame({'Mood':['Dark','Semi-Dark','Bright','Semi-Dark','Bright','Bright','Bright'],
 'Rating':['Adult','Adult','Family','Adult','Family','Adult','Adult'],
 'Class':['Dislike','Dislike','Like','Dislike','Like','Like','Like']})
```

```python
data
```

```python
data[['Mood','Rating']]
```

```python
X = pd.get_dummies(data[['Mood','Rating']])
X.head(2)
```

```python
np.array(X)
```

```python
data['Class']
```

```python
encoder = LabelEncoder()
```

```python
y = encoder.fit_transform(data['Class'])
```

```python
print(list(data['Class']))
```

```python
encoder.classes_
```

BUild Model


```python
model = Sequential([
    Dense(16,input_dim=5),
    Dense(8),
    Dense(2)
])
```

```python
model.summary()
```

Train the model

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics='accuracy')
model.fit(X,y,epochs=3)
```
